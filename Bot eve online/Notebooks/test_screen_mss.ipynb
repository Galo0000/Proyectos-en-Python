```python
import cv2
import numpy as np
import pygetwindow as gw
import pyautogui

# Función para capturar una ventana específica por su título
def capture_window(window_title):
    try:
        # Obtener la ventana por su título
        window = gw.getWindowsWithTitle(window_title)[0]
        print(window)
        # Obtener las coordenadas de la ventana
        _, _, right, bottom = window.left, window.top, window.right, window.bottom

        left = 0
        top = 23
        right -= 8
        bottom -= 8
        
        # Capturar la región de la pantalla correspondiente a la ventana
        screenshot = pyautogui.screenshot(region=(left, top, right - left, bottom - top))
        
        # Convertir la captura a una matriz numpy
        screenshot = np.array(screenshot)
        
        return screenshot
    except IndexError:
        print("La ventana especificada no fue encontrada.")
        return None

# Título de la ventana que deseas capturar
window_title = "EVE - Guileo Galo"

while True:
    # Capturar la ventana específica
    screenshot = capture_window(window_title)
    
    if screenshot is not None:
        # Mostrar la captura de la ventana en una ventana de OpenCV
        cv2.imshow("Captura de ventana", screenshot)
    
    # Esperar hasta que se presione la tecla 'q' para salir
    if cv2.waitKey(1) == ord('q'):
        break

# Liberar la ventana de OpenCV
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import pygetwindow as gw
import pyautogui
import pytesseract
from PIL import Image
import os

pytesseract.pytesseract.tesseract_cmd= r"C:\\Program Files\\Tesseract-OCR\\tesseract.exe"

# Función para capturar una ventana específica por su título
def capture_window(window_title, region):
    try:
        # Obtener la ventana por su título
        window = gw.getWindowsWithTitle(window_title)[0]
        
        # Obtener las coordenadas de la región de interés dentro de la ventana
        left, top, width, height = region
        
        # Ajustar las coordenadas a la posición relativa dentro de la ventana
        left += window.left + 8
        top += window.top + 23 + 8
        
        # Capturar la región de la pantalla correspondiente a la región de interés dentro de la ventana
        screenshot = pyautogui.screenshot(region=(left, top, width, height))
        
        # Convertir la captura a una matriz numpy
        screenshot = np.array(screenshot)
        
        return screenshot
    except IndexError:
        print("La ventana especificada no fue encontrada.")
        return None

# Título de la ventana que deseas capturar
window_title = "EVE - Guileo Galo"

# Definir la región de interés dentro de la ventana
region = (1724,325,124,18)  # (left, top, width, height)

while True:
    # Capturar la región específica de la ventana
    screenshot = capture_window(window_title, region)
    

    #screenshot = cv2.resize(screenshot, None, fx=2, fy=2)

    # Convertir la imagen a escala de grises
    gray_image = cv2.cvtColor(screenshot, cv2.COLOR_BGR2GRAY)

    # Aplicar un umbral para obtener una imagen binaria (blanco y negro)
    #_, black_white_image = cv2.threshold(gray_image, 155, 255, cv2.THRESH_BINARY)

    # Utiliza pytesseract para realizar el OCR en la imagen
    text = pytesseract.image_to_string(gray_image)

    text = text.replace('‘', '')

    # Imprime el texto extraído
    print(text) 
    
    if screenshot is not None:
        # Mostrar la captura de la ventana en una ventana de OpenCV
        cv2.imshow("Captura de ventana", gray_image)
    
    # Esperar hasta que se presione la tecla 'q' para salir
    if cv2.waitKey(1) == ord('q'):
        break

# Liberar la ventana de OpenCV
cv2.destroyAllWindows()
```
