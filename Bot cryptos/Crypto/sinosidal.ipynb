```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import talib

# Suponiendo que tienes un DataFrame llamado df con las columnas 'open', 'close', 'low', 'high' y 'volume'
# Asegrate de tener tus datos reales cargados en el DataFrame df
# Cargar datos
df = pd.read_pickle('./Datasets/df_ETHUSDT_4h.pickle')[0:200]
df['EMA3'] = talib.EMA(df['Close'].values, timeperiod=3)
# Obtener los valores mnimos y mximos de la criptomoneda
min_values = df['Low'].values
max_values = df['High'].values
ema_values = df['EMA3'].values

# Crear un rango de tiempo (x) basado en la cantidad de datos
x = np.linspace(0, len(min_values) - 1, len(min_values))


# Graficar los datos de la criptomoneda y la funcin sinusoidal
plt.figure(figsize=(30, 20))
plt.plot(x, min_values, label='Valores Mnimos de la Criptomoneda')
plt.plot(x, max_values, label='Valores Mximos de la Criptomoneda')
plt.plot(x, ema_values, label='Valores Mximos de la Criptomoneda')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()
plt.title('Aproximacin Sinusoidal de la Criptomoneda')
plt.grid(True)
plt.show()
```
